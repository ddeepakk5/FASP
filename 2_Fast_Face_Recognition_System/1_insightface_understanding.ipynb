### how to configure insightface model

```python
from insightface.app import FaceAnalysis
```

```python
# buffalo_l model
app_l = FaceAnalysis(name='buffalo_l',
                    root = 'insightface_model')
#providers = ['CPUExecutionProvider']
app_l.prepare(ctx_id=0,det_size=(640,640))
```

```python
# buffalo_ sc model
app_sc = FaceAnalysis(name='buffalo_sc',
                    root = 'insightface_model')
                    
#providers = ['CPUExecutionProvider']
app_sc.prepare(ctx_id=0,det_size=(640,640))
```

```python
img = cv2.imread("im2.jpg")
print(img)
```

### apply image to insight face model and get result

```python
results_l = app_l.get(img)
```

```python
print(results_l)
```

```python
# here we will draw the bounding box of each detected face 
img_copy = img.copy()
gender_encode = ['Female','Male']
for res in results_l:
    x1,y1,x2,y2 = res['bbox'].astype(int)
    # draw a rectangle across the face
    cv2.rectangle(img_copy,(x1,y1),(x2,y2),(0,255,0),1)

    # key points
    kps = res['kps'].astype(int)
    for k1,k2 in kps:
        cv2.circle(img_copy,(k1,k2),2,(0,255,255),-1)

    # detection score
    score ='score: {}%'.format(int(res['det_score']*100))
    cv2.putText(img_copy,score,(x1,y1),cv2.FONT_HERSHEY_TRIPLEX,0.4,(226,121,15))           # for displaying the text for age though need to tweak a little bit.

    # age and gender- for male we get 1, for female 0
    gender = gender_encode[res['gender']]
    #age = res['age']
    #age_gender = f'{gender}::{age}'
    cv2.putText(img_copy,gender,(x1,y2),cv2.FONT_HERSHEY_DUPLEX,0.4,(255,255,255))          # for displaying the text for gender




cv2.imshow('bbox',img_copy)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
